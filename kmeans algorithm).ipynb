```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

cx = np.random.rand(1600) 
cy=np.random.rand(1600)



c=np.vstack((cx,cy))
c
```

```python
plt.scatter(c[0],c[1],c="blue",s=20)
```

```python
def kmeans(k,c):
    

    


    np.random.seed(123)
    kxs=np.random.rand(k)



    np.random.seed(423)
    kys=np.random.rand(k)


    

    distances={}
    for i in range(k):
        
        distances["distance" + str(i)]=0
        
    

    
    colors=["orange","green","violet","gray","yellow","cyan"]
        
    
    

    
    ks=np.vstack((kxs,kys))
    ks.resize(k,2)
    
    
    
    plt.scatter(c[0],c[1],c="blue")
    
    
   

    cond=False
    
    
    while(cond!=True):




        ks=np.array(ks)
        ks.resize(k,2)

        cx=c[0]
        cy=c[1]



        clusters=[]


        for i in range(cx.size):
            kdist=10*6
            kind=0
            for j in range(k):
                j

                dist=math.sqrt(((cx[i] - ks[j,0]) **2) + ((cy[i] - ks[j,1]) **2))

                if dist < kdist:

                    kind=j
                    kdist=dist


            l=[cx[i], cy[i],kind]
            clusters.append(l)

        size=0
        oldk=ks.copy()
        ks=[]
        for i in range(k):
            total_value_x=0
            total_value_y=0
            color=colors[i]
            xval=[]
            yval=[]
            for cluster in clusters:
                if cluster[2]==i:



                    total_value_x+=cluster[0]
                    total_value_y+=cluster[1]
                    xval.append(cluster[0])
                    yval.append(cluster[1])


            size=len(xval)


            ks.append([total_value_x / size, total_value_y / size])
            plt.scatter(xval,yval,color=color)
            plt.scatter(total_value_x / size,total_value_y / size,color="red",s=100)




        ks=np.round(ks,12)


        for i in range(k):
                for j in range(2):

                    if ks[i][j] == oldk[i][j]:
                        cond=True
                    else:
                        cond=False
                        break

    kxval=[]
    kyval=[]
    for val in ks:

        kxval.append(val[0])
        kyval.append(val[1])
        plt.scatter(kxval,kyval,color="red",s=200)
```

```python
k=5


kmeans(k,c)
```
